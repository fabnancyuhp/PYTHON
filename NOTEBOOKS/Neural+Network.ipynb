    Les réseaux de neurones sont des algorthmes utilisés pour les problèmes de classification et de regression. Un réseau de neurone est une fonction construite à partir de poids, de biais et de fonctions de transfere. Les réseaux de neurones ont des représentations graphiques : ils sont composées de couches de neurone successives.

# Fonction de transfère
Voici des fonctions de transfère possibles:
* Sigmoide :$\sigma(z)=\frac{1}{1+\exp(-z)} $
* $Tanh$ : $\sigma(z)=\frac{\exp(z)-\exp(-z)}{\exp(z)+\exp(-z)}$
* $ReLu$ : $\sigma(z)=\max(0,z)$
* $id$ : $\sigma(z)=z$

```python
import numpy as np
x  = np.linspace(-8,8,25)
y1 = 1/(1+np.exp(-x)) #sigmoid
y2 = np.tanh(x)
y3 = np.maximum(x,0)

import matplotlib.pyplot as plt

f,ax = plt.subplots(1,3,figsize=(10,3))
ax[0].plot(x,y1,color="blue")
ax[0].set_title(r"Sigmoid: $\sigma(z)=\frac{1}{1+\exp(-z)}$")
ax[1].plot(x,y2,color="red")
ax[1].set_title(r"Tanh: $\sigma(z)=\frac{\exp(z)-\exp(-z)}{\exp(z)+\exp(-z)}$")
ax[2].plot(x,y3,color="green")
ax[2].set_title(r"ReLu: $\sigma(z)=\max(0,z)$")

plt.show()
```

* Les neux sont appellés neurones ou bien perceptrons
* L'input layer est de taille $m$, la dimension de l'espace des features $(x\in\mathbb{R}^{m})$.
* Dans un probleme de classification à k>2 classes l'output layer est de taille k. Dans ce cas, la taille de l'output est la dimension des $y^{i}$.
* Dans un probleme de classification à 2 classes, l'output layer est de dimension 1
* Dans le cas d'un probleme de regression, l'output est de dimension 1
* On notera par $S_{q}$ le nombre de neurone de la couche q.
* On notera par $b^{q}$ l'ensemble des biais associé à la couche $q$. L'input layer, c'est à dire la premiere couche n'a pas de bais. $b^{q}\in \mathbb{R}^{S_{q}}$ est un vecteur de la même taille que la couche $q$.
* On notera par $W^{q}$ l'ensemble des poids allant de la couche $q-1$ à la couche $q$. $W^{q}\in \mathbb{R}^{S_{q-1}}\times \mathbb{R}^{S_{q}}$ est une matrice.

$W$ représente l'ensemble des $W^{q}$ et $b$ représente l'ensemble des $b^{q}$.

# Entrainement d'un réseau de neurone
Comme beaucoup d'algorithme de machine learning, on cherche à minimiser la fonction de perte:
$$w^{*},b^{*}=\underset{W,b}{\mbox{argmin}}L(W,b)=\underset{W,b}{\mbox{argmin}}\frac{1}{n}\sum_{i=1}^{n}Q_{i}(W,b)+\alpha \|W\|_{2}^{2}  $$
Ou $Q_{i}(W,b)$ dépend de $(x^{i},y^{i})$. Pour résoudre ce probleme, on utilise l'un des algorithmes d'optimisation suivant:
* Adam : stochastic gradient-based optimizer
* sgd : stochastic gradient descent
* lbfgs : une méthode d'optimisation de la famille quasi-Newton

Quand l'échantillon est de grande taille on utilisera Adam. Quand l'échantillon est de petite taille on utilisera lbfgs. Le nombre d'itération maximum de ces algorithmes d'optimisation est un paramêtre sur lequel on peut jouer pour éviter l'overfiting.
Ces algorithmes ne convergent pas toujours vers un minimum global. Ils peuvent converger vers un minimums locale.

# Réseaux de neurone pour la classification
Soit $\{(x^{1},y^{1}),\dots,(x^{n},y^{n})\}$. $x^{i}\mathbb{R}^{m}$ correspond au features. $y^{i}\in\mathbb{R}^{K}$ représente la classe de $x^{i}$. Les $y^{i}$ sont les classes que l'on predit. Il y a K classes. Si la classe de $x^{i}$ est k, alors $y^{i}=[\overbrace{0,\dots,0}^{k-1},\underbrace{1}_{k},\overbrace{0,\dots,0}^{K-k}]^{t}$. On suppose que le réseau de neurone à 5 couches:
* une couche d'entré (input layer) de taille $S_{1}=m$, la dimension des $x^{i}$.
* 3 couches cachées de taille respective $S_{2},S_{3},S_{4}$
* 1 couche de sortie (output layer) de taille $S_{5}=K$
* On note par $a^{q}_{j}$ la valeur du neurone $j$ de la couche q, $q\in \{1,2,3,4\}$ 
Les neurones de le couche d'entrée ont pour valeurs les valeurs des features de $x$: $\forall j\in\{1,\dots,m\},\mbox{ }  a^{1}_{j}=x_{j}$
Alors $\forall q \in \{2,3,4,5\}, \forall j\in \{1,\dots,S_{q}\}$, on a $a^{q}_{j}=\sigma(\sum_{i=1}^{S_{q-1}}W^{q}_{i,j}a^{q-1}_{j}+b_{j}^{q})$. $\sigma$ est l'une des fonctions de transfere suivantes: Sigmoide, Tanh, ReLu, id.
$a^{4}=h_{W,b}(x)$ le vecteurs de sortie (output layer) s'écrit en fonction des matrices de poids et des vecteurs de bias :

$$h_{w,b}(x)=\sigma[W^{4}(\sigma(W^{3}\sigma[W^{2}\sigma(W^{1}x^{1}+b^{1})+b^{2}]+b^{3})+b^{4}] $$

$\begin{eqnarray*}
\forall i \in [1,\dots,m],a^{1}_{i}&=&x_{i}\\
\forall i \in [1,\dots,S_{2}], a^{2}_{i}&=&\sigma(\sum_{j=1}^{S_{1}}W_{j,i}^{2}x_{j}+b^{2}_{i})\\
a^{2}&=&\sigma(W^{2}x+b^{2})\\
\forall i \in [1,\dots,S_{3}],a^{3}_{i}&=&\sigma(\sum_{j=1}^{S_{2}}W_{j,i}^{3}a^{2}_{j}+b^{3}_{i})\\
a^{3}&=&\sigma(W^{3}a^{2}+b^{3})\\
\vdots &=&\vdots\\
\forall i \in [i,\dots,S_{q}],a_{i}^{q}&=&\sigma(\sum_{j=1}^{S_{q-1}}W_{j,i}^{q}a^{q-1}_{j}+b^{q})\\
a^{q}&=&\sigma(W^{q}a^{q-1}+b^{q})
\end{eqnarray*}$
<br>
Quand il y a 5 couches, dont une couche d'entrée, une couche de sortie et 3 couches cachées, alors la sortie s'écrit:
$$h_{w,b}(x)=\sigma(W^{5}\sigma(W^{4}\sigma(W^{3}\sigma(W^{2}x+b^{2})+b^{3})+b^{4})+b^{5}) $$
Quand $S_{5}=K$, le vecteur de probabilité de sortie est de la forme $h_{w,b}(x)=(h_{w,b}(x)_{1},\dots,h_{w,b}(x)_{K})$

$\forall i\in\{1,\dots,n\},h_{W,b}(x^{i})$ un vecteur à K dimension. On va calculer la vraisemblance:
$$L((x_{1},y_{1})\dots,(x^{n},y^{n}))=\prod_{i=1}^{n}\prod_{k=1}^{K}h_{W,b}(x^{i})_{k}^{y^{i}_{k}}\times (1-h_{W,b}(x^{i})_{k})^{1-y^{i}_{k}}$$
Donc la logvraissemblance s'écrit:
$$\log(L((x_{1},y_{1})\dots,(x^{n},y^{n})))=\sum_{i=1}^{n}\sum_{k=1}^{K}y^{i}_{k}\log(h_{W,b}(x^{i})_{k})+(1-y^{i}_{k})\log(1-h_{W,b}(x^{i})_{k}) $$
On va donc chercher à minimiser la fonction de perte/cout suivantes:
$$J(W,b)=-\frac{1}{n}\sum_{i=1}^{n}\sum_{k=1}^{K}y^{i}_{k}\log(h_{W,b}(x^{i})_{k})+(1-y^{i}_{k})\log(1-h_{W,b}(x^{i})_{k})+\lambda\|W\|_{2}^{2}$$
ou $\lambda\|W\|_{2}^{2}$ permet d'éviter l'overfiting ($\lambda>0$).
Ensuite l'étape d'entrainement du réseau de neurone commence. A l'aide d'un algorithme d'optimisation on cherche
$W^{*},b^{*}=\underset{W,b}{\mbox{argmin}}J(W,b)$. Les algorithmes d'optimisation possibles sont adam, SGD,lbfgs. Le nombre maximal d'itération de l'algorithme d'optimisation est un paramêtre du réseau de neurone. 

# prediction d'un $x^{new}$
    
Dans le cadre d'une classification, les $h_{W^{*},b^{*}}(x^{new})_{k}$ représentent la probabilité que $x^{new}$ soit dans la classe $k$. Donc 
$\mbox{classe}(x^{new})=\underset{k\in \{1,\dots,K\}}{\mbox{argmax}}h_{W^{*},b^{*}}(x^{new})_{k}$

# Réseau de neurone pour la régression

Soit $\{(x^{1},y^{1}),\dots,(x^{n},y^{n})\}$. $x^{i}\mathbb{R}^{m}$ correspond au features. $y^{i}\in\mathbb{R}$ représente est la variable dépandante de $x^{i}$ que l'on cherhce à prédir. On suppose que le réseau de neurone à 5 couches:
* une couche d'entré (input layer) de taille $S_{1}=m$, la dimension des $x^{i}$.
* 3 couches cachées de taille respective $S_{2},S_{3},S_{4}$
* 1 couche de sortie (output layer) de taille $S_{5}=1$ puisque $y\in\mathbb{R}$. $b^{5}\in \mathbb{R}$.
* On note par $a^{q}_{j}$ la valeur du neurone $j$ de la couche q, $q\in \{1,2,3,4\}$ 
Les neurones de le couche d'entrée ont pour valeurs les composantes du feature $x$: $\forall j\in\{1,\dots,m\},\mbox{ }  a^{1}_{j}=x_{j}$
Alors $\forall q \in \{2,3,4\}, \forall j\in \{1,\dots,S_{q}\}$, on a $a^{q}_{j}=\sigma(\sum_{i=1}^{S_{q-1}}W^{q}_{i,j}a^{q-1}_{j}+b_{j}^{q})$. $\sigma$ est l'une des fonctions de transfere suivantes: Sigmoide, Tanh, ReLu, id.
$a^{5}=h_{W,b}(x)$ le vecteurs de sortie (output layer) s'écrit en fonction des matrices de poids et des vecteurs de biais :

$$h_{w,b}(x)=W^{5}\sigma(W^{4}\sigma[W^{3}\sigma(W^{2}x+b^{2})+b^{3}]+b^{4})+b^{5}$$

Contrairement à la classification, on n'utilise pas de fonction de transfére dans la couche de sortie. La couche de sortie a une seule dimention et le biais $b^{5}$ a une seule dimention. 

On cherche à minimiser la fonction de perte/cout suivante:

$$J(W,b)=\frac{1}{n}\sum_{i=1}^{n}\left(h_{w,b}(x^{i})-y^{i}\right)^{2}+\lambda\|W\|^{2} $$
ou $\lambda\|W\|_{2}^{2}$ permet d'éviter l'overfiting ($\lambda>0$). Ensuite l'étape d'entrainement du réseau de neurone commence. A l'aide d'un algorithme d'optimisation on cherche
$W^{*},b^{*}=\underset{W,b}{\mbox{argmin}}J(W,b)$. Les algorithmes d'optimisation possibles sont adam, SGD,lbfgs. Le nombre maximal d'itération de l'algorithme d'optimisation est un paramêtre du réseau de neurone. 

La prediction de $x^{new}$ se calcule à partir de $W^{*},b^{*}$:
$\hat{y^{new}}=h_{w^{*},b^{*}}(x^{new})$.

### exemple classification sur les données iris
Les données iris ont 4 features. Il y  a 3 classes possibles.
Dans l'exemple ci-dessous, on va implémenter un réseau de neurole type MLP à 2 couche caché de 3 neuornes et 2 neurones

On import les données irir

```python
from sklearn.datasets import load_iris

iris = load_iris()

X = iris['data']
Y = iris['target']

from sklearn.model_selection import train_test_split


X_train, X_test, Y_train, Y_test = train_test_split()
```

On divise en échantillon d'apprentissage et de test

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.25)
```

On procéde à un scallage des données.

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(X_train)

X_train_scal = scaler.transform(X_train)
X_test_scal = scaler.transform(X_test)
```

On appelle le réseau de neurone avec 2 couche caché de 2 neurones et 3 neurones (hidden_layer_sizes=(2,3))

```python
from sklearn.neural_network import MLPClassifier

clf = MLPClassifier(hidden_layer_sizes=(2,3),activation='logistic',solver='lbfgs')

clf.fit(X_train_scal,Y_train)
```

On fait une prédiction sur les test:

```python
clf.predict(X_test_scal)
```

On cherche les poids de ce réseaut de neurone. Ceci se fait avec l'objet coefs_:

```python
clf.coefs_
```

Les poids qui vont l'input à la premier couche caché sont:

```python
clf.coefs_[0]
```

Les poids qui vont de la deuxième couche caché a l'output sont:

```python
clf.coefs_[2]
```

On a 3 classes. Donc l'objet predict_proba doit retourner des vecteurs de probabilité de taille 3: 

```python
clf.predict_proba(X_test_scal)[0:4]
```

On regarde les biais. On doit avoir 2 bias à la premiere couche caché, 3 bias à la deuxième couche caché et 3 biais pour la couche de sortie. On utilise l'objet intercepts_:

```python
clf.intercepts_
```

### Exemple de réseau de neurone pour la régression : prédire la température critique de supraconducteur
Les données proviennent du site https://archive.ics.uci.edu/ml/machine-learning-databases/00464/

```python
import pandas as pd
supracond = pd.read_csv("/home/fabien/Bureau/donnees projet/superconduct/train.csv")
```

On reformate les données:

```python
list_feature = list(supracond)
target = ['critical_temp']
list_feature = [o for o in list_feature if o not in ['critical_temp','number_of_elements']]
ADUM = supracond[['number_of_elements']].astype(str)
ADUMBIS = pd.get_dummies(ADUM)
X =  supracond[list_feature]
X = pd.concat([ADUMBIS,X],axis=1)
Y = supracond[target]
```

On divise en échantillon d'apprentissage et de test:

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.20)
```

On scale les données:

```python
from sklearn.preprocessing import StandardScaler
scale = StandardScaler()
scale.fit(X_train)
X_train_scale = scale.transform(X_train)
X_test_scale = scale.transform(X_test)
```

Ici on implémente un réseau de neurone à 3 couches cachées de 25 neurones chacunes:

```python
from sklearn.neural_network import MLPRegressor

NNR = MLPRegressor(hidden_layer_sizes=(25,25,25),activation='relu',solver='lbfgs',\
                   alpha=500,max_iter=5000,learning_rate='adaptive')

#NNR = MLPRegressor(hidden_layer_sizes=(15,15),activation='relu',solver='lbfgs',\
                   #alpha=500,max_iter=5000,learning_rate='adaptive')
```

Etape d'apprentissage:

```python
NNR.fit(X_train_scale,Y_train.values.ravel())
```

Calcule de la métrique r2:

```python
from sklearn.metrics import r2_score
Y_pred = NNR.predict(X_test_scale)
r2_score(Y_test,Y_pred)
```
